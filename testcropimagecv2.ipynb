```python
import cv2
import matplotlib.pyplot as plt
import easyocr
```

```python
img = cv2.imread("Dataset\datasetsuhu (25).jpg")
```

```python
cropped = img[65:1200,0:1920]
```

```python
plt.title('Cropped Image')
plt.imshow(cropped)
```

```python
reader=easyocr.Reader(['en'], gpu=True) #English
result=(reader.readtext(cropped))
result1=result[0][1]
print(result1[-4:])
```
